```python
# Numeric columns impute with mean
num_cols = ['Avg VTAT','Avg CTAT','Booking Value','Ride Distance','Driver Ratings','Customer Rating']
for col in num_cols:
    df[col] = df[col].fillna(df[col].mean())

# Categorical columns impute with mode
cat_cols = ['Payment Method']
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

##### Categorical columns impute with mode

```python
# Categorical columns impute with mode
cat_cols = ['Payment Method']
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.isnull().sum()
```

##### Check duplicate count

```python
print("Duplicate rows:", df.duplicated().sum())

# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
# Convert Date and Time properly
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
# Combine Date + Time into one datetime column
df['Pickup_datetime'] = pd.to_datetime(df['Date'].astype(str) + ' ' + df['Time'].astype(str), errors='coerce')

# Extract useful time features
df['Year'] = df['Pickup_datetime'].dt.year
df['Month'] = df['Pickup_datetime'].dt.month
df['Day'] = df['Pickup_datetime'].dt.day
df['Hour'] = df['Pickup_datetime'].dt.hour
df['DayOfWeek'] = df['Pickup_datetime'].dt.day_name()
```

```python
df['Booking ID'] = df['Booking ID'].astype(str)
df['Customer ID'] = df['Customer ID'].astype(str)
```

```python
print(df.info())
```

##### Convert object → category

```python
cat_cols = [
    'Time','Booking ID','Booking Status','Customer ID',
    'Vehicle Type','Pickup Location','Drop Location',
    'Payment Method','DayOfWeek'
]

for col in cat_cols:
    df[col] = df[col].astype('category')
```

```python

plt.figure(figsize=(10,5))
sns.countplot(x='Hour', data=df, palette="viridis")
plt.title("Peak Booking Hours", fontsize=16)
plt.xlabel("Hour of Day")
plt.ylabel("Number of Bookings")
plt.show()
```

```python
#####1. Booking Trends
```

```python
plt.figure(figsize=(12,5))
df['DayOfWeek'].value_counts().reindex(
    ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
).plot(kind='bar', color='teal')
plt.title("Bookings by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Number of Bookings")
plt.show()
```

Higher booking demand on weekends (Saturday & Sunday)

Moderate demand on weekdays (Monday–Friday)

Business Insight: Increase driver availability on weekends to meet higher demand

#### 2. Peak Hours

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Hour', data=df, palette="viridis")
plt.title("Peak Booking Hours")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Bookings")
plt.show()
```

```python
Morning (8–10 AM) and evening (6–9 PM) are the busiest hours

Very few late-night bookings

Business Insight: Apply surge pricing during peak hours to maximize revenue
```

#### 3. Vehicle Type Analysis

```python
plt.figure(figsize=(8,5))
df['Vehicle Type'].value_counts().plot(kind='bar', color='orange')
plt.title("Vehicle Type Distribution")
plt.xlabel("Vehicle Type")
plt.ylabel("Number of Bookings")
plt.show()
```

Sedans and Hatchbacks are the most frequently booked vehicles

SUVs are less common

Business Insight: Maintain a larger fleet of Sedans/Hatchbacks to match demand

#### 4. Payment Method

```python
plt.figure(figsize=(6,6))
df['Payment Method'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("Preferred Payment Methods")
plt.show()
```

```python
UPI/Wallets are the most popular payment options

Cash payments are relatively low

Business Insight: Promote digital payments further with cashback/discounts
```

#### 5. Ride Distance vs Booking Value

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Ride Distance', y='Booking Value', alpha=0.3)
plt.title("Ride Distance vs Booking Value")
plt.xlabel("Ride Distance (km)")
plt.ylabel("Booking Value (₹)")
plt.show()
```

```python
Booking value increases with ride distance (expected positive correlation)

Some outliers indicate high-value corporate or premium rides

Business Insight: Introduce special packages for long-distance rides
```

#### 6. Ratings

```python
fig, ax = plt.subplots(1,2, figsize=(12,5))
sns.histplot(df['Driver Ratings'], bins=10, kde=True, ax=ax[0], color="blue")
ax[0].set_title("Driver Ratings Distribution")

sns.histplot(df['Customer Rating'], bins=10, kde=True, ax=ax[1], color="green")
ax[1].set_title("Customer Ratings Distribution")
plt.show()
```

```python
Majority of ratings are 4 stars or higher → indicates good service quality

Few low ratings exist → service inconsistency in some cases

Business Insight: Monitor low-rated drivers/customers and provide training or feedback
```
